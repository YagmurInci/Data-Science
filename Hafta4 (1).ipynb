```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#pythonda uyarıları kapatır
import warnings
warnings.filterwarnings("ignore")
```

```python
veri=pd.read_csv(r"C:\Users\USER\Downloads\38833FF26BA1D.UnigramPreview_g9c9v27vpyspw!App\olimpiyatlar.csv")
veri.head(15)
```

```python
# tablonun teknik bir özetini verir
veri.info()
```

```python
veri.rename(columns={'ID':'id',
'Name':'isim',
'Sex':'cinsiyet',
'Age':'yas',
'Height':'boy',
'Weight':'kilo',
'Team':'takim',
'NOC':'kod',
'Games':'oyunlar',
'Year':'yil',
'Season':'sezon',
'City':'sehir',
'Sport':'spor',
'Event':'etkinlik',
'Medal':'madalya'},inplace=True)
```

```python
# id ve oyunlar sütunlarını(axis=1) siler 
veri=veri.drop(['id','oyunlar'],axis=1)
```

```python
veri.head(3)
```

```python
essiz_etkinlik=pd.unique(veri.etkinlik)
print(f"eşsiz etkinlik kimlik sayısı={len(essiz_etkinlik)}")
print(essiz_etkinlik[:20])
```

```python
# boy ve kilo sütununda bulunan kayıp veriyi etkinlik ortalamasına göre doldurucaz
veri_gecici=veri.copy()
degisenler_listesi=["boy","kilo"]
for e in essiz_etkinlik:
    etkinlik_filtre=veri_gecici.etkinlik==e
    veri_filtreli=veri_gecici[etkinlik_filtre]
    for s in degisenler_listesi:
        ortalama =np.round(np.mean(veri_filtreli[s]),2)
        #print("etkinlik=",e," ",s,"'nin ortalaması",ortalama)
        if not np.isnan(ortalama): #ortalama bir sayıysa # not yerine ~ de koyabilirsin
            veri_filtreli[s]=veri_filtreli[s].fillna(ortalama)
        else: # ortalama sayı değilse
            genel_ort=np.round(np.mean(veri_gecici[s]),2)
            veri_filtreli[s]=veri_filtreli[s].fillna(genel_ort)
    veri_gecici[etkinlik_filtre]=veri_filtreli

#veri=veri_gecici(copy)
#veri.info()
        
```

```python
veri=veri_gecici.copy()
veri.info()
#veri_gecici.info()
```

```python
yas_ortalamasi=np.round(np.mean(veri.yas),2)
veri.yas.fillna(yas_ortalamasi,inplace=True)
```
