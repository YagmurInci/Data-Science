```python
sns.lmplot(x='Beygir',y='Fiyat_TL',
           data=df,
           hue='Yakit_Tipi',
          fit_reg=False, # regresyon çizgisini gizlemek için
          height=6, #grafik yüksekliği
          aspect=1.5,# genişlik / yğkseklik oranı,
          palette='Set2',# renk paleti
          scatter_kws={'s':150,'alpha':0.5} # nokta boyutu: s değeri ,şeffaflık: alpha
          )
plt.title('Beygir Gücü ve Fiyat ilişkisi Tablosu')
#web scrapping internetten veri çekme
```

```python
# yakıt tipine göre beygir gücü dağılımıtek çizer.
plt.figure(figsize=(15,5))
sns.violinplot(x='Yakit_Tipi',y='Beygir',data=df,palette='muted',inner=None,linewidth=2)

sns.swarmplot(x='Yakit_Tipi',y='Beygir',data=df,color='purple',alpha=0.5,size=9)
```

```python
stats_df=df[['Marka','Model','Segment','Beygir','Tork']]
stats_df.head()
```

```python
df.head()
```

```python
print(f"Şekil:{stats_df.shape}")
```

```python
print(f"Orjinal şekil:{df.shape}")
```

```python
#Melt işlemi Beygir ve Tork sütunlarını "özellik" adlı tek bir sütuna indiriyoruz
melt_df=pd.melt(stats_df,
                id_vars=["Marka","Model","Segment"],
                value_vars=["Beygir","Tork"],
                var_name="Ozellik",
                value_name="Deger"
               )
melt_df.head()
```

```python
print(melt_df.shape)
```

```python
#beygir ve tork karşılaştırması
plt.figure(figsize=(12,7))
sns.swarmplot(x='Ozellik',y='Deger',data=melt_df,
             hue='Segment',
             palette='viridis',
             size=4,
             alpha=0.7
             )
```

```python
numeric_df=df.select_dtypes(include=['float64','int64'])
corr=numeric_df.corr() # korelasyon hesaplama (+1 ve -1 aralığında dönüştürme)
#print(corr)
sns.heatmap(corr,annot=True, # kutucuklar içerisinde değerleri görmek
           fmt=".2f", #ondalık sayıların gösterimi
            linewidth=0.5, # bloklar arası boşluk
            cbar_kws={"shrink":0.8} # renk çubuğu ayarı
           )
```
