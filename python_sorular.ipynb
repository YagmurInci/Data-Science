```python
kullanici1={
    'ad':'yagmur',
    'soyad':'inci',
    'uzmanlık':['front-end']
}
kullanici2={
    'ad':'fatma',
    'soyad':'inci',
    'uzmanlık':['tasarim','yazilim']
}
kullanici3={
     'ad':'toprak',
    'soyad':'inci',
    'uzmanlık':['front-end']
}
```

```python
#yagmur inci uzmnlık alani
kullanici1['uzmanlık']
```

```python
#front-end lerin ismini yazdır
kullanici_listesi=[kullanici1,kullanici2,kullanici3]
for kullanici in kullanici_listesi:
    if kullanici['uzmanlık']==['front-end']:#if 'front-ed'in kullanici.get('uzmanlik'): dmk dahadgru aksi hale fronted dısınds bseydaha varsakbuletmez
       print(kullanici['ad'])
```

```python
#toprakın uzmanlıgına yazılım ekle
kullanici3['uzmanlık'].append('yazılım')
print(kullanici_listesi)
```

```python
#birden fazla uzmanlık olanları dondur (hint:lenght)
for kullanici in kullanici_listesi:
    if len(kullanici['uzmanlık']) > 1:
        print(kullanici)
```

```python
kullanici_yas=[19,38,4]
#zip kullanarak iki listeyi birlestir ve yas 30 danaz olanlari dondur
for kullanici,yas in zip(kullanici_listesi,kullanici_yas):
    if yas<30:
        print(kullanici)
```

```python
#deger isimli degiskene atanan syinn asal olma durumunu döndur
deger=7
x=deger-1
while x>1:
   if deger%x==0:
      print("{} sayisi asal degildir".format(deger))
      break
   else:
      x-=1
else :      
    print("{} sayisi asaldir".format(deger))
```
