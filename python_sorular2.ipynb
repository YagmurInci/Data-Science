```python
# soru:4=asagıdaki fonku liste tipinde bir parametre alacak ve sadece str 
#tipindeki degerleri filter ve lambda ifadelerini kullanarak filtreleyecek bicimde olustur
```

```python
def str_filtrele(liste):
    sonuc=[]
    for x in liste:
        if type(x)==str :
            sonuc.append(x)
        else:
            pass
    return sonuc
```

```python
str_filtrele([1,2,"asd","a",5,True])
```

```python
#filterla cozum
def str_filtrele_2(liste):
    return [*filter(lambda x:x if type(x)==str else None , liste)]
```

```python
str_filtrele_2([1,2,"asd","a",5,True])
```

```python
#soru:5=fonku 1 parametre alacak (sadece sayi iceren liste) ve map ve lambda kullanarak 6 sifir atacak bicimde olusturun
```

```python
def alti_sifir_sil(liste):
    return [*map(lambda x:int(x/1000000),liste)]
     
```

```python
alti_sifir_sil([1000000,90000000,15000000])
```

```python
def alti_sifir_silv2(liste):
    sonuc=[]
    for x in liste:
        sonuc.append(int(x/1000000))
    return sonuc
```

```python
alti_sifir_silv2([1000000,90000000,15000000])
```

```python
def zaman_verisi_al():
    saat= input("saat giriniz:")
    if saat.isdigit():
        saat=int(saat)
        if ((saat>=0 )and (saat<24)):
             dakika= input("dakika giriniz:")
             if dakika.isdigit():
                 dakika=int(dakika)
                 if ((dakika>=0) and (dakika<60)):
                     return("giris yapilan zaman {}:{}".format(saat,dakika))
                 else:
                     return "girilen dakika araligi yanlis"
             else:
                 return "girilen dakika yanlis veri tipinde"
        else:
            return "girilen saat araligi yanlis"
    else:
        return "girilen saat yanlis veri tipinde"
        
```
