## Görev 3: Eksik Veri Doldurma (25 puan) 
1. director sütunu için:  
• Eksik değerleri "Unknown Director" ile doldurun 
2. cast sütunu için:  
• Eksik değerleri "No Cast Info" ile doldurun 
3. country sütunu için:  
• Eksik değerleri en sık geçen ülke ile doldurun (mode) 
4. date_added ve duration sütunları için:  
• Eksik değerleri içeren satırları silin 
5. rating sütunu için:  
• Movie ve TV Show için ayrı ayrı en sık geçen rating ile doldurun

### 1. director sütunu için:  ###    
• Eksik değerleri "Unknown Director" ile dolduma 

```python
df["director"].isnull().sum() 
```

```python
df.director.fillna("Unknown Director",inplace=True) # değişiklikleri kaydet (inplace=True)
df.head()
```

```python
df.director.isnull().sum() 
```

### 2. cast sütunu için:  ###    
• Eksik değerleri "No Cast Info" ile dolduma 

```python
df.cast.isnull().sum()
```

```python
df.cast.fillna("No Cast Info",inplace=True)
df.head()
```

```python
df.cast.isnull().sum()
```

### 3. country sütunu için:  ###    
• Eksik değerleri en sık geçen ülke ile doldurun (mode) 

```python
df["country"] = df["country"].fillna(df["country"].mode()[0])
df.head()
```

### 4. date_added ve duration sütunları için:  ###   
• Eksik değerleri içeren satırları silin

```python
df[["date_added", "duration"]].isnull().sum()
```

```python
df = df.dropna(subset=["date_added", "duration"])
df[["date_added", "duration"]].isnull().sum()
```

### 5. rating sütunu için:  ###   
• Movie ve TV Show için ayrı ayrı en sık geçen rating ile doldurun

```python
df.head()
```

```python
movie_mode = df[df["type"] == "Movie"]["rating"].mode()[0]
tv_mode = df[df["type"] == "TV Show"]["rating"].mode()[0]
```

```python
df.loc[(df["type"] == "Movie") & (df["rating"].isnull()), "rating"] = movie_mode
df.loc[(df["type"] == "TV Show") & (df["rating"].isnull()), "rating"] = tv_mode
```

## Görev 4: Veri Görselleştirme  
a) Movie vs TV Show Dağılımı 
b) Yıllara Göre Eklenen İçerik Sayısı 
c) En Popüler 10 Ülke 
d) Release Year Dağılımı (Histogram)

f) Türkiyede gösterilere türler

### 1. Movie vs TV Show Dağılımı 

```python
movie_count = df["type"].value_counts()["Movie"]
tvshow_count = df["type"].value_counts()["TV Show"]

plt.figure()
plt.bar(["Movie", "TV Show"], [movie_count, tvshow_count],color="pink")
plt.title("Movie vs TV Show Dağılımı")
plt.xlabel("içerik türü")
plt.ylabel("içerik sayısı")
plt.xticks(rotation=0)
plt.show()
```

### 2. Yıllara Göre Eklenen İçerik Sayısı 

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
# Başındaki ve sonundaki boşlukları temizler
df["date_added"] = df["date_added"].str.strip()

# date_added'i datetime formatına çevirir
df["date_added"] = pd.to_datetime(df["date_added"], errors="ignore")
year_counts = df["date_added"].dt.year.value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.bar(year_counts.index, year_counts.values)
plt.title("Yıllara Göre Eklenen İçerik Sayısı")
plt.xlabel("Yıl")
plt.ylabel("İçerik Sayısı")

plt.show()

```

### 3. En Popüler 10 Ülke 

```python
import matplotlib.pyplot as plt
def plotBar(column, n=10):
    veri_sayma = df[column].dropna().value_counts().head(n)
    plt.figure(figsize=(10,5))
    plt.bar(veri_sayma.index, veri_sayma.values, color="pink")
    plt.xticks(rotation=60)
    plt.xlabel("Ülke")
    plt.ylabel("İçerik Sayısı")
    plt.title("En Popüler 10 Ülke")
    plt.show()
plotBar("country", n=10)
```

### 4.Release Year Dağılımı (Histogram) 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.hist(df["release_year"], bins=15)
plt.title("Release Year Dağılımı")
plt.xlabel("Yıl")
plt.ylabel("İçerik Sayısı")
plt.show()
```

### 5. Movie Sürelerinin Dağılımı (dakika olarak)

```python
# Sadece Movie olanları seçelim
movies = df[df['type'] == 'Movie']

# Süreleri dakikaya çevirelim
movies['duration_minutes'] = movies['duration'].str.replace(" min","").astype(int)

movies['duration_minutes'].plot(kind='hist', bins=30, color='orange')
plt.title("Movie Sürelerinin Dağılımı")
plt.xlabel("Dakika")
plt.ylabel("Frekans")
plt.show()
# Movie Sürelerinin Dağılımı (dakika olarak) (e)
```

### 6. Türkiyede gösterilen türler

```python
import matplotlib.pyplot as plt
turkey_filter = df[df["country"].str.contains("Turkey", na=False)]
turler = turkey_filter["listed_in"].str.split(",").explode().value_counts()
plt.figure(figsize=(15,10))
plt.bar(turler.index,turler.values)
plt.title("Türkiye’de Gösterilen İçerik Türleri")
plt.xlabel("Tür")
plt.ylabel("İçerik Sayısı")
plt.xticks(rotation=65)
plt.show()
```
